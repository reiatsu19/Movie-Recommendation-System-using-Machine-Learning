```python
#Importing neccessary libraries
import numpy as np
import pandas as pd
import ast
import pickle
```

```python
#Reading the dataset from download path
path = '/drive/MyDrive/Colab Notebooks/tmdb_5000_movies.csv'
movies = pd.read_csv(path)
```

```python
#Displaying the all attributes in the dataset
movies
```

```python
#Viewing the attributes in a more readable form
movies.info()
```

```python
#Selecting only the headers which are relevant for the project
movies = movies[['id', 'title', 'overview', 'genres', 'keywords']]
```

```python
#Viewing the dataset post-selection
movies.head()
```

```python
#Viewing the current state of the selected data regarding null data
movies.info()
```

```python
#Viewing which columns contain null values
movies.isnull().sum()
```

```python
#Dropping the rows which contain null values in its columns and applying the changes to the same dataframe instead of creating a copy
movies.dropna(inplace=True)
```

```python
#Viewing the current state of the selected data after data cleaning
movies.info()
movies.isnull().sum()
```

```python
#Takes string of text which represents list of dictionaries and returns list of values related to 'name' in each dictionary
def conversion(text):
    t = []
    for i in ast.literal_eval(text):
        t.append(i['name'])
    return t
```

```python
movies['genres'] = movies['genres'].apply(conversion)
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(conversion)
movies.head(1)
```

```python
#Removes spaces in empty strings
def removal(t):
    t2 = []
    for i in t:
        t2.append(i.replace(" ", ""))
    return t2
```

```python
movies['genres'] = movies['genres'].apply(removal)
movies['keywords'] = movies['keywords'].apply(removal)
movies.head(1)
```

```python
#Splitting each sentence into list of words
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head(1)
```

```python
#Combining the main movie similarity criteria into a single group
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']
```

```python
#Creating a new dataframe where the combined columns are dropped
movies2 = movies.drop(columns=['overview','genres','keywords'])
```

```python
#Join the new group created earlier with the new dataframe
movies2['tags'] = movies2['tags'].apply(lambda x: " ".join(x))
movies2.head(1)
```

```python
#Extrating features from the text data where minimum document frequency is 2
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english', min_df=2)  # Set min_df to 2
```

```python
#Creating a matrix or vector of word counts for each movie
vectors = cv.fit_transform(movies2['tags']).toarray()
```

```python
#Using cosine similarity on the vectors
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
# Define a function to recommend movies based on similarity to input movie
def recommend(movie):
    index = movies2[movies2['title'] == movie].index
    if len(index) == 0:
        print('Sorry, movie not found.')
    else:
        index = index[0]
        distance = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
        for i in distance[1:6]:
            print(movies2.iloc[i[0]].title)
```

```python
#Deplying the recommender model on movie Avatar
recommend('Avatar')
```

```python
#Preparing the dataframes into a pickle file for use in PyCharm during GUI construction
pickle.dump(movies2,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
